# E. Дискретизация признаков

1.	Проведите дискретизацию признаков, используя KBinsDiscretizer, на следующее количество диапазонов:
age - 5
creatinine_phosphokinase - 4
ejection_fraction - 6
platelets - 10
serum_creatinine - 7
serum_sodium - 4


```python
from sklearn.preprocessing import KBinsDiscretizer

n_bins = [5, 4, 6, 10, 7, 4]

discretizer = KBinsDiscretizer(n_bins=n_bins, encode='ordinal', strategy='uniform')
data_discretized = discretizer.fit_transform(data)

fig, axs = plt.subplots(2, 3, figsize=(15, 10))

feature_names = ['age', 'creatinine_phosphokinase', 'ejection_fraction', 'platelets', 'serum_creatinine', 'serum_sodium']

# Построение гистограмм для каждого признака
for i in range(2):
    for j in range(3):
        feature_index = i * 3 + j
        axs[i, j].hist(data_discretized[:, feature_index], bins=n_bins[feature_index])
        axs[i, j].set_title(f'{feature_names[feature_index]}')
        axs[i, j].set_xlabel('Диапазоны')
        axs[i, j].set_ylabel('Частота')
        axs[i, j].set_xticks(range(n_bins[feature_index]))

plt.show()
```

Мы разделили значения на n частей, потом построили гистограммы. Что получилось можно посмотреть выше

```python
for i in range(6):
    print(f"\nДиапозон для признака {i+1}")
    for j in range(n_bins[i]):
        print(f"{discretizer.bin_edges_[i][j]} ",end='')
```
